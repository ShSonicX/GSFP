## Аппроксимация геометрическим средним (Gentel)

```python
# аппроксимация корзины опционов геометрическим средним
def geometric_average_approximation_basket_call(S, K, T, w, r, rho, sigma):
  F = S * np.exp(r * T)
  a = w * F / np.sum(w * F)

  m_tilde = np.log(np.sum(w * F)) - 0.5 * np.sum(a * sigma ** 2 * T)
  v_tilde = np.sqrt((a * sigma).T @ rho @ (a * sigma) * T)

  K_star =  K - np.sum(w * F) + np.exp(m_tilde + 0.5 * v_tilde ** 2)

  d1 = (m_tilde - np.log(K_star) + v_tilde ** 2) / v_tilde
  d2 = d1 - v_tilde

  basket_call = np.exp(-r * T) * (np.exp(m_tilde + 0.5 * v_tilde ** 2) * norm.cdf(d1) - K_star * norm.cdf(d2))

  return basket_call
geometric_average_approximation_basket_call(S, K, T, w, r, rho, sigma)
```

## Метод условного математического ожидания (Beisser)

```python
def get_z_star(x, S, K, r, q, sigma, T, w, rho):
    res = 0
    a = S * w * sigma
    sqrt_T = np.sqrt(T)
    sigmaz = np.sqrt(a.T @ rho @ a)
    for i in range(0, len(w)):
        m = np.sum(a * rho[i]) / sigmaz**2 * sqrt_T
        exp_S_Z = S[i] * np.exp(sigma[i] * m * x - 1/2 * sigma[i]**2 * m**2 * sigmaz**2 + (r - q[i]) * T)
        res += w[i] * exp_S_Z
    return (res - K)**2

def Beisser_approximation_basket_call(S, K, T, w, r, q, sigma):
  option = 0
  a = S * w * sigma
  sigmaz = np.sqrt(a.T @ rho @ a)
  sqrt_T = np.sqrt(T)

  z_star = minimize(get_z_star, 0, args=(S, K, r, q, sigma, T, w, rho)).x[0]
  for i in range(0, len(w)):
    m = np.sum(a * rho[i]) / sigmaz**2 * sqrt_T
    sigma_star = sigma[i] * m * sigmaz / np.sqrt(T)
    K_star = S[i] * np.exp(sigma[i] * m * z_star - 1/2 * sigma[i]**2 * m**2 * sigmaz**2 + (r - q[i]) * T)
    S_star = S[i] * np.exp((r - q[i] - 1/2 * sigma_star**2) * T + sigma_star * np.sqrt(T) / sigmaz * z_star)
    d1 = (np.log(S[i] / K_star) + (r - q[i] + 1/2 * sigma_star**2) * T) / (sigma_star * np.sqrt(T))
    d2 = d1 - sigma_star * np.sqrt(T)
    option += w[i] * (np.exp(-q[i] * T) * S[i] * norm.cdf(d1) - np.exp(-r * T) * K_star * norm.cdf(d2))
    #print(f"d1 = {d1}, d2 = {d2}, z_star = {z_star}, K_star = {K_star}, S_star = {S_star}, sigma_star = {sigma_star}")
  return option
Beisser_approximation_basket_call(S, K, T, w, r, q, sigma)
```

### Метод Джу через разложения в ряд Тейлора (Ju)

Статья [(Ju, 2002)](https://drive.google.com/file/d/1JJuZhRiCUGf6euyfju1Dn5aRP46-3iry/view?usp=drive_link)

```python
def basket_option_price_approximation(asset_prices, strike_price, time_to_maturity, volatilities, weights, correlation_coefficient):
    # Формирование корреляционной матрицы
    correlation_matrix = np.full((len(asset_prices), len(asset_prices)), correlation_coefficient) + np.eye(len(asset_prices)) * (1 - correlation_coefficient)

    # Расчет дисперсии корзины активов с учетом взвешенной волатильности
    vol_squared = volatilities ** 2
    weighted_variance = np.dot(weights, np.dot(correlation_matrix * np.outer(volatilities, volatilities), weights))
    basket_volatility = np.sqrt(weighted_variance)  # Общая волатильность корзины

    # Определение стоимости корзины активов на текущий момент времени
    basket_value = np.sum(weights * asset_prices)

    # Вычисление параметров d1 и d2 для модели Блэка-Шоулза
    d1 = (np.log(basket_value / strike_price) + (0.5 * basket_volatility**2) * time_to_maturity) / (basket_volatility * np.sqrt(time_to_maturity))
    d2 = d1 - basket_volatility * np.sqrt(time_to_maturity)

    # Использование формулы Блэка-Шоулза для расчета цены call опциона на корзину активов
    option_price = basket_value * norm.cdf(d1) - strike_price * norm.cdf(d2)

    return option_price

# Пример использования функции
option_price = basket_option_price_approximation(S, K, T, sigma, w, rho)
option_price
```

## Метод моментов (Log-normal moment matching, Levy’s matching)

```python
def Levy_approximation_basket_call(S, K, T, w, r, rho, sigma):
  F = S * np.exp(r * T)
  M = np.dot(w, F)
  V2 = 0
  for i in range(0, len(w)):
    for j in range(0, len(w)):
      V2 += w[i] * w[j] * F[i] * F[j] * np.exp(sigma[i] * sigma[j] * rho[i][j] * T)
  m = 2 * np.log(M) - 1/2 * np.log(V2)
  v = np.sqrt(np.log(V2) - 2 * np.log(M))
  d1 = (m - np.log(K) + v**2) / v
  d2 = d1 - v
  basket_call = np.exp(-r * T) * (M * norm.cdf(d1) - K * norm.cdf(d2))
  return basket_call
Levy_approximation_basket_call(S, K, T, w, r, rho, sigma)
```

## Аппроксимация через обратное геометрическое распределение (MP-RG)

```python
def Milevsky_Posner_RG_approximation_basket_call(S, K, T, w, r, rho, sigma):
  F = S * np.exp(r * T)
  M = np.dot(w, F)
  V2 = 0
  for i in range(0, len(w)):
    for j in range(0, len(w)):
      V2 += w[i] * w[j] * F[i] * F[j] * np.exp(sigma[i] * sigma[j] * rho[i][j] * T)
  alpha = (2 * V2  - M**2) / (V2 - M**2)
  beta = (V2 - M**2) / (V2 * M)
  g1 = gamma.cdf(1/K, a = alpha - 1, scale = beta)
  g2 = gamma.cdf(1/K, a = alpha, scale = beta)
  basket_call = np.exp(-r * T) * (M * g1 - K * g2)
  return basket_call
Milevsky_Posner_RG_approximation_basket_call(S, K, T, w, r, rho, sigma)
```

### Метод моментов высоких порядков (MP-4M)

Статья [(Posner and Milevsky, 1998)](https://https://drive.google.com/file/d/16f5dRiA_HD9jhm-esUT_6gGoArk9Nc2u/view?usp=drive_link)

```python
import numpy as np
from scipy.stats import norm

def Milevsky_Posner_4M_approximation_basket_call(S, K, T, w, r, rho, sigma):
    """
    Метод моментов высоких порядков (MP-4M) для оценки корзинного колл-опциона.
    
    Параметры:
    S - начальная стоимость корзины (массив или список)
    K - цена страйка
    T - время до истечения опциона
    w - веса для корзины (массив или список)
    r - безрисковая ставка
    rho - коэффициент корреляции между активами
    sigma - волатильности для каждого актива в корзине (массив или список)
    
    Возвращает:
    basket_call - приближенная цена корзинного колл-опциона
    """
    
    N = len(S)
    
    # Моменты первого порядка (среднее)
    mu = np.dot(w, S)
    
    # Преобразуем списки в массивы NumPy для операций с элементами
    w = np.array(w)
    sigma = np.array(sigma)
    
    # Моменты второго порядка (дисперсия)
    sigma2 = 0
    for i in range(N):
        for j in range(N):
            sigma2 += w[i] * w[j] * rho[i, j] * sigma[i] * sigma[j]
    
    # Моменты третьего порядка
    sigma3 = 0
    for i in range(N):
        for j in range(i, N):
            sigma3 += 2 * w[i] * w[j] * rho[i, j] * sigma[i] * sigma[j]
    
    # Моменты четвертого порядка
    sigma4 = 0
    for i in range(N):
        for j in range(i, N):
            sigma4 += 2 * w[i] * w[j] * (sigma[i] * sigma[j] + rho[i, j] * (sigma[i] ** 2 + sigma[j] ** 2))
    
    # Рассчитываем элементы модели
    d1 = (np.log(mu / K) + (r + 0.5 * sigma2) * T) / np.sqrt(sigma2 * T)
    d2 = d1 - np.sqrt(sigma2 * T)
    
    # Цена корзинного колл-опциона
    basket_call = np.exp(-r * T) * (mu * norm.cdf(d1) - K * norm.cdf(d2))
    
    return basket_call

# Данные
T = 5
r = 0
K = 100
N = 4

rho0 = 0.5
rho  = np.ones((N, N)) * rho0 + np.diag(np.ones(N)) * (1 - rho0)
S0   = 100
S    = np.ones(N) * S0
F    = np.ones(N) * S0 * np.exp(r * T)
w    = np.ones(N) / N
q    = np.zeros(N)
sigma = np.ones(N) * 0.4

# Получаем цену корзинного колл-опциона
basket_call = Milevsky_Posner_4M_approximation_basket_call(S, K, T, w, r, rho, sigma)
print(f'Цена корзинного колл-опциона: {basket_call:.2f}')


# ответ = 27.98
```

## Метод Монте-Карло

```python
def MC_basket_call(S, K, T, w, r, q, rho, sigma, N_mc):
  n = len(w)
  COV = sigma.T * rho * sigma
  W = np.random.multivariate_normal(np.zeros(n), COV, size=N_mc) * np.sqrt(T)
  B_T = np.zeros(N_mc)
  S0 = 100
  for k in range(0, n):
      B_T += S[k] * np.exp((r - q[k] - sigma[k]**2 / 2) * T + W[:,k]) * w[k]
  basket_call = np.mean(np.maximum(B_T - K, 0)) * np.exp(-r * T)
  return basket_call
N_mc = int(10**6)
MC_basket_call(S, K, T, w, r, q, rho, sigma, N_mc)
```

```python
def MC_basket_call_with_std(S, K, T, w, r, q, rho, sigma, N_mc):
  n = len(w)
  COV = sigma.T * rho * sigma
  W = np.random.multivariate_normal(np.zeros(n), COV, size=N_mc) * np.sqrt(T)
  B_T = np.zeros(N_mc)
  S0 = 100
  for k in range(0, n):
      B_T += S[k] * np.exp((r - q[k] - sigma[k]**2 / 2) * T + W[:,k]) * w[k]
  payoff = np.maximum(B_T - K, 0)
  mean = np.mean(payoff) * np.exp(-r * T)
  std = np.std(payoff) * np.exp(-r * T)
  return call_price_mc, std

def CI_MC(price, price_std, N_mc):
    lo = price - norm.ppf(0.975)*price_std / np.sqrt(N_mc)    # lower bound of confidence interval for the price
    hi = price + norm.ppf(0.975)*price_std / np.sqrt(N_mc)    # upper bound of confidence interval for the price
    print(f"mean call basket price = {price}, ci: {lo} < price < {hi} with 95% prob")
```

```python
def monte_carlo_analysis(F, K, T, r, sigma, rho, w, n_sims=10**4, n_steps=5000, n_repeats=100, confidence=0.95):
    """
    Выполняет Монте-Карло моделирование для оценки цен европейских опционов, 
    вычисляет доверительные интервалы и визуализирует результаты.

    Аргументы:
    F - начальные цены активов
    K - страйк
    T - время до погашения
    r - безрисковая ставка
    sigma - волатильности активов
    rho - коэффициент корреляции
    w - веса портфеля
    n_sims - количество симуляций
    n_steps - количество временных шагов
    n_repeats - количество повторений моделирования
    confidence - уровень доверия для интервалов
    """
    n_assets = len(F)  # Количество активов
    dt = T / n_steps  # Шаг по времени
    cov_matrix = np.outer(sigma, sigma) * rho  # Ковариационная матрица

    call_prices = []  # Список для цен call-опционов
    put_prices = []  # Список для цен put-опционов

    for _ in range(n_repeats):  # Повторяем моделирование
        # Инициализация цен активов с явным указанием типа данных float
        S = np.tile(F, (n_sims, 1)).astype(float)
        for _ in range(n_steps):  # Моделируем траектории цен
            dW = np.random.multivariate_normal(np.zeros(n_assets), cov_matrix, size=n_sims) * np.sqrt(dt)
            S += r * S * dt + S * dW  # Учет дрейфа и диффузии
        
        # Расчет стоимости портфеля и выплат опционов
        portfolio_values = S @ w
        call_payoffs = np.maximum(portfolio_values - K, 0)
        put_payoffs = np.maximum(K - portfolio_values, 0)
        discount_factor = np.exp(-r * T)
        call_prices.append(discount_factor * np.mean(call_payoffs))
        put_prices.append(discount_factor * np.mean(put_payoffs))

    # Преобразуем списки в массивы
    call_prices = np.array(call_prices)
    put_prices = np.array(put_prices)

    # Функция для расчета доверительных интервалов
    def calculate_ci(data):
        mean = np.mean(data)  # Среднее значение
        margin = norm.ppf(0.5 + confidence / 2) * (np.std(data, ddof=1) / np.sqrt(len(data)))
        return mean, mean - margin, mean + margin

    # Расчет доверительных интервалов
    call_mean, call_ci_low, call_ci_high = calculate_ci(call_prices)
    put_mean, put_ci_low, put_ci_high = calculate_ci(put_prices)

    # Построение гистограмм
    plt.figure(figsize=(12, 5))
    for i, (data, title, xlabel, mean, ci_low, ci_high) in enumerate([
        (call_prices, "Распределение цен Call-опционов", "Цена Call-опциона", call_mean, call_ci_low, call_ci_high),
        (put_prices, "Распределение цен Put-опционов", "Цена Put-опциона", put_mean, put_ci_low, put_ci_high),
    ]):
        plt.subplot(1, 2, i + 1)
        plt.hist(data, bins=20, alpha=0.6, color='green' if i == 0 else 'purple')  # Цвета для Call и Put
        plt.axvline(mean, color='orange', linestyle='--', label="Среднее")
        plt.axvline(ci_low, color='red', linestyle='--', label="95% ДИ")
        plt.axvline(ci_high, color='red', linestyle='--')
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel("Частота")
        plt.legend()
    plt.tight_layout()
    plt.show()

    # Проверка нормальности распределений
    for data, label in [(call_prices, "Call"), (put_prices, "Put")]:
        p_value = shapiro(data)[1]  # Тест Шапиро-Уилка
        result = "нормально распределены" if p_value > 0.05 else "не нормально распределены"
        print(f"{label} Цены: Данные {result} (p-value = {p_value:.4f})")

    return call_prices, put_prices
```
